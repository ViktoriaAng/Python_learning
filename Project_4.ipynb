```python
# Кількість респондентів
total_respondents = df.ResponseId.nunique()
print(total_respondents)
```

```python
#кількість людей, що відповіли на всі 10 обов'язкових запитань
df_clean = df.dropna(subset=['ResponseId', 'MainBranch', 'Age', 'Employment', 'RemoteWork', 'Check', 'CodingActivities', 'EdLevel', 'LearnCode', 'LearnCodeOnline'])
df_clean.shape[0]
```

```python
# Обчислення мір центральної тенденції
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0]

print(f"Середній досвід: {mean_exp}, Медіанний досвід: {median_exp}, Мода: {mode_exp}")
```

```python
#Кількість респондентів, які працюють віддалено
remote_respondents = df[df['RemoteWork'] == 'Remote'].shape[0]
print(remote_respondents)
```

```python
# Підрахунок респондентів, які програмують на Python
python_devs = df[df['LanguageHaveWorkedWith'].str.contains('Python','python', na=False)].shape[0]
python_percentage = (python_devs / total_respondents) * 100
print(python_devs)
print(f"Відсоток респондентів, які програмують на Python: {python_percentage:.2f}%")

```

```python
# Підрахунок респондентів, які навчалися програмуванню за допомогою онлайн-курсів
online_learning_respondents = df[df['LearnCode'].str.contains('Online Courses', na=False)].shape[0]
print(online_learning_respondents)
```

```python
# Групування за країнами і обчислення середньої та медіанної компенсації
python_salary = df[df['LanguageHaveWorkedWith'].str.contains('Python','python', na=False)]
python_salary['ConvertedCompYearly'] = pd.to_numeric(python_salary['ConvertedCompYearly'], errors='coerce')

salary_stats_by_country = python_salary.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).reset_index()

print(salary_stats_by_country)
```

```python
# Відбір 5 рівнів освіти з найбільшою компенсацією
top_5_highest_paid = df.nlargest(5, 'ConvertedCompYearly')
education_levels = top_5_highest_paid[['ConvertedCompYearly', 'EdLevel']]
print(education_levels)
```

```python
# Підрахунок респондентів, які програмують на Python в кожній віковій категорії
age_groups = df.groupby('Age')
total_age_counts = age_groups.size()
python_age_counts = age_groups['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python', na=False).sum())
python_percentage_by_age = (python_age_counts / total_age_counts) * 100
print(python_percentage_by_age)
```

```python
# Підрахування найбільш поширених індустрій в 75 перцентилі.
percentile_75 = df['ConvertedCompYearly'].quantile(0.75)
high_salary_remote = df[(df['ConvertedCompYearly'] > percentile_75) & (df['RemoteWork'] == 'Remote')]
industry_counts = high_salary_remote['Employment'].value_counts()
print(industry_counts.head(5))
```
